CREACIÓN DE REGISTROS EN TABLA DE AIRTABLE (POST)

```python
# FUNCION PREPARACION DE DATOS Y CREACIÓN DE REGISTROS EN AIRTABLE A PARTIR DE DF 


def creacion_registros(df):
    
  
    i = 0 # Inicializa el contador de filas del df original (df_indices)
    j = 1 # Inicializa el contador de paginas de 10 filas cada una (df_paginas_indices)

# Crea un dict para almacenar dict de 10 filas cada uno hasta recorrer todas las filas del df original (df_indices)
    paginas_indices = {}
    

    while i < df.shape[0]:   # iniciliza un bucle while, ejecutable mientras i sea menor a la cantidad de filas 
        indices = {"records"  : [],
                   "typecast" : True}
    
        filas = df.iloc[i:i+10, : ]

        for _, fila in filas.iterrows():
            fila_dict = {"fields" : fila.to_dict()}  
            indices["records"].append(fila_dict)

            paginas_indices[f"indices_{j}"] = indices
        
        i += 10 # Incrementa el contador de filas del df original (df_indices) en 10 filas
        j += 1 # Incrementa el contador de paginas del df_paginas_indices en 1 pagina, para pasar a la siguiente pagina de 10 filas

   
    print("-"*120)

    for key, pagina in paginas_indices.items():  # Iterea sobre el diccionario de paginas_indices, posteando cada pagina en la API de Airtable mediante un bucle for
        response = requests.post(url = url_endpoint, json = pagina, headers = headers) # Postea los registros a la tabla de AirTable
 
          
    print(f"response: {response.status_code}")

    print(f"endpoint: {response.url}")
    
    total_registros = sum(len(pagina["records"]) for pagina in paginas_indices.values()) # Suma el total de registros en cada pagina para darme el número total de registros cargados
    
    print(f"Total de registros cargados: {total_registros}")

    print("-"*120)
        
    pprint(paginas_indices)
    
    sleep(0.5)
           
creacion_registros(df_indices_urls)
```

VISUALIZACIÓN DE REGISTROS DE TABLA DE AIRTABLE (GET)

```python
# Endpoint
endpoint = url_endpoint

params = {"fields"                : None, 
          "maxRecords"            : None, 
          "pageSize"              : None,
          "returnFieldsByFieldId" : None}

print(endpoint)

pprint(params, sort_dicts = False)
```

```python
# FUNCIÓN LISTADO DE REGISTROS EN AIRTABLE
 
def listado_registros(endpoint):
    
    # Crea una lista vacia para almacenar los registros de la tabla de AirTable
    registros_extendidos = list()
    
    # Bucle para recorrer todos los registros de la tabla de AirTable
    while params.get("offset") != None or len(registros_extendidos) == 0:
    
        response = requests.get(url = endpoint, headers = headers, params = params)
        
        
        
        registros = response.json() # Guarda en variable el json de la respuesta 

        # Si no hay registros en la respuesta, termina el bucle y me indica que la tabla no teine registros
        if not registros["records"]:
            
            print("No hay registros en esta tabla de AirTable")
            break 
            
        # Offset para fragmentar los registros obtenidos, sorteandola paginación de AirTable en caso de contener más de los 100 registros por tabla permitidos por esta API
        offset = registros.get("offset") 
        
        params["offset"] = offset
        
        registros_extendidos.extend(registros["records"])
                       
        print("-"*120)
        
        print("Registros obtenidos de tabla de AirTable")
        print("Cantidad de registros en la tabla Lista Indices:")
        print(len(registros_extendidos))
        
        print(f"response: {response.status_code}")
        
        print(f"endpoint: {response.url}")
        

        pprint(response.json(), sort_dicts = False)
        
        sleep(0.5)
    
    records_id = [x["id"] for x in response.json()["records"]]# ID de cada fila de la tabla de AirTable
    registros_extendidos = pd.json_normalize(registros_extendidos)  # Normaliza los registros extendidos     
    df_registros = pd.DataFrame(registros_extendidos) # Guarda los registros en un DataFrame
   
    print("-"*120)
    return df_registros, records_id
    
    
              
df_registros_AirTable, records_id = listado_registros(url_endpoint) # Crea variable del df de los registros de la tabla obtenido con la función, para ser utilizados fuera de la función

df_indice_url_limpio = df_registros_AirTable.drop(columns=["id", "createdTime"], axis = 1) # Elimina las columnas de id y createdTime del df de registros de la tabla de AirTable
nombres_columnas = ["Índice" , "URL"] # Nuevos nombres para columnas del df
df_indice_url_limpio.columns = nombres_columnas # Renombra las columnas del df de registros de la tabla de AirTable para mayor legibilidad

df_indice_url_limpio.to_csv("./data/Airtable/df_indice_url_limpio.csv", index=False) # Guarda df obtenido de los registros de la Tabla de Airtable en archivo CSV, en caso de necesitar exportar los mismos
```

ACTUALIZACIÓN DE REGISTROS EN LA TABLA DE AIRTABLE (PATCH)

```python
# MODIFICACIÓN DE REGRISTROS A MODIFICAR Y CARGA DE MODIFICACIONES EN AIRTABLE (PATCH)

i = 0 # Inicializa contador de filas
j = 0 # Inicializa contador de paginas 

registros_modificados = [] # Crea lista vacia para almacenar registros modificados en cada iteracion del bucle

while i < len(records_id):
    # Crea diccionario "url_completa", que contiene "subdiccionario records"
    # En diccionario "records" se crea un subdiccionario "fields"
    # Itera sobre "records" poupulando "id" con los valores de "records_id" creado anteriormente al descargar tabla con método GET
    url_completa = {"records" : [{"id" : x, "fields" : {}} for x in records_id[i:i+10]]}
    
    
    if len(url_completa["records"]) == 0: # Termina el bucle si no hay mas registros sobre los que iterar
        break
        
    # Convierte el df_registros_Airtable en dicicionario, tomando "id" como índice y "fields.Value" como valor
    df_dict = df_registros_AirTable.set_index('id')['fields.Value'].to_dict()

    # Bucle que itera sobre los registros de "records" en "url_completa"
    #Verifica si 
    for rec in range(len(url_completa["records"])): 
        record_id = url_completa["records"][rec]["id"] # Popula "id" en records con los valores de records_id
        if record_id in df_dict: # Verifica si el valor de "id" está en el diccionario df_dict
            url_completa["records"][rec]["fields"]["Value"] = "https://es.finance.yahoo.com" + df_dict[record_id] # Agrega el comienzo de la URL antes de los registros de "Values" para cada correspondiente "record_id"

    response = requests.patch(url = endpoint, json = url_completa, headers = headers)
    
    registros_modificados.append(response.json()) # Guarda registros modificados en cada iteración del bucle
    
    i += 10    
    j += 1
    
    sleep(0.5)

print(f"Se han modificado {len(records_id)} registros en la tabla de AirTable")

print(f"response: {response.status_code}")

print(f"endpoint: {response.url}")

print("-"*120)

pprint(registros_modificados)

print("-"*120)

```
